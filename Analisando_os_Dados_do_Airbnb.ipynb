## Análise dos Dados
Esta etapa tem como objetivo permitir um entendimento de como os dados estão estruturados.


**Dicionário das variáveis**

* `id` - número de id gerado para identificar o imóvel
* `name` - nome da propriedade anunciada
* `host_id` - número de id do proprietário (anfitrião) da propriedade
* `host_name` - Nome do anfitrião

* `longitude` - coordenada da longitude da propriedade

* `price` - preço para alugar o imóvel
* `minimum_nights` - quantidade mínima de noites para reservar

* `calculated_host_listings_count` - quantidade de imóveis do mesmo anfitrião

### **Há *outliers* presentes?**

A distribuição do histograma indica presença de *outliers* ou seja, valores atípicos. Tomando como exemplo as variáveis `price`, `minimum_nights` e `calculated_host_listings_count` é possível perceber essa distorção quando comparadas com outras variáveis.

Em seguida, serão apresentadas duas formas de detecção de *outliers*.

Aqui será utilizado o resumo estatístico das variáveis numéricas por meio do método `describe()`

```
# resumo estatístico das variáveis numéricas
### df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].describe()
df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].describe()
```

Olhando o resumo estatístico acima, podemos confirmar algumas hipóteses como:

#### BOX PLOT ####

Boxplot para minimum_nights

```
# minumum_nights
# listar a quantidade de valores acima de 21 dias para minimum_nights
df.minimum_nights.plot(kind='box', vert=False, figsize=(16,4))
plt.show()

# ver quantidade de valores acima de 21 dias para minimum_nights
print("minimum_nights com valores acima de 21 dias:")
print("{} entradas".format(len(df[df.minimum_nights > 21])))
print("{:.4f}%".format((len(df[df.minimum_nights > 21]) / df.shape[0])*100))
```

Boxplot para Price

```
#price
df.price.plot(kind='box', vert=False, figsize=(16,4))
plt.show()

# ver quantidade de valores acima de 1200 para price
print("\nprice com valores acima de 1200")
print("{} entradas".format(len(df[df.price > 1200])))
print("{:.4f}%".format((len(df[df.price > 1200]) / df.shape[0])*100))
```

#### **Histogram sem Outliers** ####
Após a comprovação da existência de *outliers* nas variáveis `price` e `minimum_nights`, o próximo passo será a limpeza de seus *DataFrame* e plotar histograma novamente.

```
# remoção dos outliers, através da criação de um novo DataFrame
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 1200].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 21].index, axis=0, inplace=True)

# plotar o histogram
df_clean.hist(bins=15, figsize=(20,15),  color='#014742');
```

#Observação
Em função da existência de valores *outliers* em minimum_nights e price, os novos pontos de corte para estas colunas passaram ser de 21 dias e 1000$, respectivamente.

### **Qual a média dos preços do aluguel com e sem outliers**?


```
# Média da coluna price (valor do aluguel)
##round(df_clean['price'].mean(),2)

price_no_outliers = round(df_clean['price'].mean(),2)
price_outliers = round(df['price'].mean(),2)

print("Média do aluguel com outlires é de",price_outliers)
print("Média do aluguel sem outlires é de",price_no_outliers)

```

### **Qual a correlação existente entre as variáveis**

Correlação significa a existência de relação entre duas coisas. No contexto dessa análise, busca-se a relação ou semelhança entre duas variáveis.

Essa relação pode ser medida, e é função do coeficiente de correlação estabelecer qual a intensidade dela. 

De forma a identificar as correlações existentes entre as variáveis de interesse serão efetuadas as ações abaixo: 

* Criar uma matriz de correlação
* Gerar um *heatmap* a partir dessa matriz, usando a biblioteca `seaborn`

```
# criar uma matriz de correlação
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']].corr()
display(corr)
```

```
# mostrar a matriz de correlação
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

O Heatmap acima permite identificar a correlação de 69% entre a quantidade de reviews e quantidade de reviews por mês.

### **Qual o tipo de imóvel mais alugado no Airbnb?**
A coluna da variável room_type indica o tipo de locação que está anunciada no Airbnb. 

A seguir, será demosntrada a quantidade de ocorrências de cada tipo de aluguel, usando o método value_counts().

```
# mostrar a quantidade de cada tipo de imóvel disponível
df_clean.room_type.value_counts()
```

```
# Crie um Plot que mostre o número de veículos pertencentes a cada marca
sns.set_style("whitegrid")
g = sns.catplot(y="room_type", data=df_clean, kind="count", palette="Reds_r", height=5, aspect=1.5)
g.ax.set_title("Tipo de Imóvel",fontdict={'size':18})
g.ax.xaxis.set_label_text("Quantidade",fontdict= {'size':16})
g.ax.yaxis.set_label_text("Tipo",fontdict= {'size':16})
plt.show()
```

```
# mostrar a porcentagem de cada tipo de imóvel disponível
df_clean.room_type.value_counts() / df_clean.shape[0] * 100
```

De acordo com os dados acima é possível identificar que as maiores prefeências são por alugar apenas um quarto(58,11%) e imóveis inteiros(37,97%)
